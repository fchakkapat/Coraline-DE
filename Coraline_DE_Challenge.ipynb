````python
#Connecting to the PostgreSQL Database by using psycopg2 module

import psycopg2

#***Set variables here***
db_name = 'challenge'
db_user = 'root'
db_pw = 'DataEngineer_2023'

try:
      conn = psycopg2.connect(
          database = db_name,
          user = db_user,
          password = db_pw
      )
      cursor = conn.cursor()
      print("Connection established successfully!")
except Exception as e:
      print("Error connecting to the database:", e)
````

````python
#Create function to create table in database
tb_name1 = 'food_sales'

def create_tables():
  commands = (
  f'''
    DROP TABLE IF EXISTS {tb_name1}
  ''',
  f'''
    CREATE TABLE IF NOT EXISTS {tb_name1} (
          ID VARCHAR(255) PRIMARY KEY,
          Date DATE NOT NULL,
          Region VARCHAR(255) NOT NULL,
          City VARCHAR(255) NOT NULL,
          Category VARCHAR(255) NOT NULL,
          Product VARCHAR(255) NOT NULL,
          Qty INT NOT NULL,
          UnitPrice FLOAT NOT NULL,
          TotalPrice FLOAT NOT NULL )
  ''')
  try:
      cursor = conn.cursor()
      for command in commands :
          cursor.execute(command)
      conn.commit()
      print(f"Table {tb_name1} created successfully")
      cursor.close()
  except Exception as error:
      print("Error: %s" % error)
      conn.rollback()
      cursor.close()

create_tables()
````

````python
#Create function to insert Dataframe into table
def insert_df():
  from psycopg2 import extras

  insert_query = f'''
  INSERT INTO {tb_name1} (
    ID,
    Date,
    Region,
    City,
    Category,
    Product,
    Qty,
    UnitPrice,
    TotalPrice)
  VALUES %s
  '''
  print(f"*START INSERT DATA FROM `FoodSales` Format `sheet` INTO `{tb_name1}` Format Postgresql database*")
  try:
    cursor = conn.cursor()
    extras.execute_values(cur=cursor,sql=insert_query,argslist=df_clean.values)
    conn.commit()
    print("the dataframe is inserted >> Success")
    cursor.close()
  except (Exception, psycopg2.DatabaseError) as error:
    print("Error: %s" % error)
    conn.rollback()
    cursor.close()

insert_df()
````

````python
#Check for inserted data in postgresql
cursor = conn.cursor()
query = "SELECT * FROM food_sales"
cursor.execute(query)
result = cursor.fetchall()
columns = ['ID', 'Date', 'Region', 'City', 'Category', 'Product', 'Qty', 'UnitPrice', 'TotalPrice']
pg_df = pd.DataFrame(result, columns=columns)
pg_df
````

> As above, **Data has been moved from sheet to Postgresql completely**.

````python
# Close connection
cursor.close()
conn.close()
````

## Assignment No.2  
- **SQL script**  
create a table: cat_reg, using data from Assignment No.1 .

````python
#Connecting to the PostgreSQL Database
try:
    conn = psycopg2.connect(
        database = db_name,
        user = db_user,
        password = db_pw
    )
    cursor = conn.cursor()
    print("Connection established successfully!")
except Exception as e:
    print("Error connecting to the database:", e)
````

````python
#Create function to create table name : cat_reg
tb_name2 = 'cat_reg'
def create_tables2():

  table_creation2 = (
    f"""
      DROP TABLE IF EXISTS {tb_name2}
    """,
    f"""
      CREATE TABLE IF NOT EXISTS {tb_name2} (
            Category VARCHAR(255) NOT NULL,
            East INT NOT NULL,
            West INT NOT NULL,
            "Grand Total" INT NOT NULL )
    """)
  try:
      cursor = conn.cursor()
      for tb_c2 in table_creation2 :
        cursor.execute(tb_c2)
      conn.commit()
      print(f"Table {tb_name2} created successfully")
      cursor.close()
  except Exception as error:
      print("Error: %s" % error)
      conn.rollback()
      cursor.close()

create_tables2()
````

> **table cat_reg** is cross table format,or Excel pivot table.

````python
#Query for check condition pivot table before insert data

query1 = '''
SELECT
  Category,
  SUM(CASE WHEN Region = 'East' THEN TotalPrice END) AS East,
  SUM(CASE WHEN Region = 'West' THEN TotalPrice END) AS West,
  SUM(TotalPrice) AS "Grand_Total"
FROM food_sales
GROUP BY Category
ORDER BY Category;
'''
cursor.execute(query1)
print(cursor.fetchall())
````

````python
#Create function to insert data into cat_reg table
def insert_data2():
  insert_pv = f'''
  INSERT INTO {tb_name2}
  SELECT
    Category,
    SUM(CASE WHEN Region = 'East' THEN TotalPrice END) AS East,
    SUM(CASE WHEN Region = 'West' THEN TotalPrice END) AS West,
    SUM(TotalPrice) AS "Grand_Total"
  FROM food_sales
  GROUP BY Category
  ORDER BY Category;
  '''
  print(f"*START CREATE DATA FROM `FoodSales` TO `{tb_name2}`*")
  try:
    cursor = conn.cursor()
    cursor.execute(insert_pv)
    conn.commit()
    print("the dataframe is inserted >> Success")
    cursor.close()
  except (Exception, psycopg2.DatabaseError) as error:
    print("Error: %s" % error)
    conn.rollback()
    cursor.close()

insert_data2()
````

````python
#Check result
cursor = conn.cursor()
cursor.execute('SELECT * FROM cat_reg')
tb = cursor.fetchall()
tb_df = pd.DataFrame(tb,columns=['Category','East','West','Grand Total'])
tb_df
````

> As above, **The total price table has been achieved from using the data from Assignment No.1** .
